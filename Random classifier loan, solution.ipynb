```python
df=df.drop('Loan_ID', axis = 1)
```

```python
# Replacing emplty values in Gender, Married, Dependents, Self_employed, Loan Amount, Loan Amount Term and Credt History
df['LoanAmount'].fillna(df['LoanAmount'].median(),inplace=True)
df['Gender'].fillna('unknown',inplace=True)
df['Married'].fillna('unknown',inplace=True)
df['Dependents'].fillna('0',inplace=True)
df['Self_Employed'].fillna('unknown',inplace=True)
df['Loan_Amount_Term'].fillna('0',inplace=True)
df['Credit_History'].fillna('0',inplace=True)
```

# Removing all the outliers

```python
# Remvong the Outliers from Applicant Income
max_p=df['ApplicantIncome'].quantile(0.99)
max_p
min_p=df['ApplicantIncome'].quantile(0.0)
min_p
df=df[(df['ApplicantIncome']<max_p)&(df['ApplicantIncome']>min_p)] 
```

```python
# Remvong the Outliers from Co Applicant Income
max_p=df['CoapplicantIncome'].quantile(0.99)
max_p
min_p=df['CoapplicantIncome'].quantile(0.01)
min_p
df=df[(df['CoapplicantIncome']<max_p)&(df['CoapplicantIncome']>min_p)] 
```

```python
# Remvong the Outliers from Co Applicant Income
max_p=df['LoanAmount'].quantile(0.99)
max_p
min_p=df['LoanAmount'].quantile(0.01)
min_p
df=df[(df['LoanAmount']<max_p)&(df['LoanAmount']>min_p)] 
```

```python
#Use stricter outlier method
Interquantile range
```

```python
Q1 = df['CoapplicantIncome'].quantile(0.25)
Q3 = df['CoapplicantIncome'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['CoapplicantIncome'] >= lower_bound) & (df['CoapplicantIncome'] <= upper_bound)]
```

```python
Q1 = df['ApplicantIncome'].quantile(0.25)
Q3 = df['ApplicantIncome'].quantile(0.75)
IQR = Q3 - Q1
lower_boundA = Q1 - 1.5 * IQR
upper_boundA = Q3 + 1.5 * IQR
df = df[(df['ApplicantIncome'] >= lower_boundA) & (df['ApplicantIncome'] <= upper_boundA)]
```

# Data Prepartation and Training Loan Approval Prediction Model using Encoder

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df.dtypes
```

```python
df.shape
```

```python
# Converting categorical columns to numerical column
textcol = ['Gender', 'Married','Dependents', 'Education','Self_Employed','Property_Area']
Label = LabelEncoder()
for col in textcol:
    df[col]=Label.fit_transform(df[col])
```

```python
df.head()
```

# Training the dataset

```python
# Preparing the data into train and test dataset

x= df.drop('Loan_Status', axis=1)
y = df['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=40)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
# Predict our x_test

y_pred = model.predict(x_test)
y_pred
```

```python
model.score(x,y)
```
